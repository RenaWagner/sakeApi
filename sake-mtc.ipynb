# Sake MTC

```python
import requests
from bs4 import BeautifulSoup
```

## single sake

```python
url = "https://www.mtcsake.com/sake/amabuki-daiginjo"
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser').body
print(soup.prettify())
```

```python
product_info = soup.find(id="productDetails")
print(product_info.prettify())
```

```python
product_title = product_info.find(class_="product-title").text
product_title
```

```python
product_details = product_info.find(class_="product-excerpt")
print(product_details.prettify())
```

```python
sake_type_raw = product_details.contents[1].get_text()
sake_type = sake_type_raw[7:]
sake_type
```

```python
sake_rice_raw = product_details.contents[2].get_text()
sake_rice = sake_rice_raw[6:]
sake_rice
```

```python
sake_rice_milling_raw = product_details.contents[3].get_text()
sake_rice_milling = sake_rice_milling_raw[22:]
sake_rice_milling
```

```python
def get_sake_details(product_info):
    product_title = product_info.find(class_="product-title").text
    product_details = product_info.find(class_="product-excerpt")
    sake_type_raw = product_details.contents[1].get_text()
    sake_type = sake_type_raw[7:]
    sake_rice_raw = product_details.contents[2].get_text()
    sake_rice = sake_rice_raw[6:]
    sake_rice_milling_raw = product_details.contents[3].get_text()
    sake_rice_milling = sake_rice_milling_raw[22:]
    sake_details = {"title": product_title,
                    "type": sake_type,
                    "rice": sake_rice,
                    "polish_percentage": sake_rice_milling,
                    "alcohol": "",
                   }
    return sake_details
get_sake_details(product_info)
```

## Get all URLs

```python
global_page_url = "https://www.mtcsake.com/sake"
global_req = requests.get(global_page_url)
global_soup = BeautifulSoup(global_req.text, 'html.parser').body
print(global_soup.prettify())
```

```python
main_content = global_soup.find("div", {"id": "productList"})
print(main_content.prettify())
```

```python
link = []
base_url = "https://www.mtcsake.com"
for a in main_content.find_all("a", href=True):
    if a.text:
        specific_url=a['href']
        each_url= f'https://www.mtcsake.com{specific_url}'
        link.append(each_url)
link
```
